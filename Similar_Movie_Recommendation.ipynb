Import libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Import Dataset into PD

```python
#Import dataset

movie_dataset = pd.read_csv('/content/movies.csv')
```

```python
movie_dataset.head()
```

```python
movie_dataset.shape
```

#Feature extraction

```python
list_of_features = ['genres', 'keywords', 'overview', 'tagline', 'title', 'cast', 'director']
```

```python
for feature in list_of_features:
  movie_dataset[feature] = movie_dataset[feature].fillna('')
```

```python
feature_dataset = movie_dataset['genres'] +' '+ movie_dataset['keywords']+' '+ movie_dataset['overview'] +' '+ movie_dataset['tagline']+' '+ movie_dataset['title'] +' '+ movie_dataset['cast'] +' '+ movie_dataset['director']
```

```python
feature_dataset.head()
```

```python
feature_dataset.shape
```

```python
vector = TfidfVectorizer()
```

```python
feature_vector = vector.fit_transform(feature_dataset)
```

```python
print(feature_vector)
```

```python
feature_vector.shape
```

#Generate Cosine Similarity

```python
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

#Find Similar Movies

i.Get similar movie

```python
movie_input = input('enter a movie :')
```

```python
all_title = movie_dataset['title'].tolist()
```

```python
print(all_title)
```

```python
close_movies = difflib.get_close_matches(movie_input, all_title)
```

```python
print(close_movies[0])
```

```python
close_movie = close_movies[0]
```

```python
close_movie_index = movie_dataset[movie_dataset.title == close_movie]['index'].values[0]
```

```python
print(close_movie_index)
```

#Get top movies module

```python
similarity_list = list(enumerate(similarity[close_movie_index]))
```

```python
print(similarity_list)
```

```python
sorted_similarity_list = sorted(similarity_list, key=lambda x: x[1], reverse=True)
```

```python
print(sorted_similarity_list)
```

```python
i=0

for movie in sorted_similarity_list:
  movie_index = movie[0]
  if i<20:
    movie_name = movie_dataset[movie_dataset.index==movie_index]['title'].values[0]
    i+=1
    print(movie_name)
```

```python
#get close movie based on input:

movie_input = input('enter a movie :')
all_title = movie_dataset['title'].tolist()
close_movies = difflib.get_close_matches(movie_input, all_title)
close_movie = close_movies[0]
print('close movie identified for given input: '+ close_movie)
close_movie_index = movie_dataset[movie_dataset.title == close_movie]['index'].values[0]



#Identity top 30movies for close movie:
similarity_list = list(enumerate(similarity[close_movie_index]))
sorted_similarity_list = sorted(similarity_list, key=lambda x: x[1], reverse=True)


i=0

for movie in sorted_similarity_list:
  movie_index = movie[0]
  if i<20:
    movie_name = movie_dataset[movie_dataset.index==movie_index]['title'].values[0]
    i+=1
    print(movie_name)
```
